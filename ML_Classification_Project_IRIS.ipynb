```python
# Import Frameworks 

import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
iris = load_iris()
```

# Checking the working data

```python
dir(iris)
```

```python
iris.feature_names
```

```python
X = iris.data
X
```

```python
y = iris.target
y
```

```python
iris.target_names # ['setosa', 'versicolor', 'virginica']
```

```python
# train test split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
y_train
```

# Logistic Regression

```python
# create the model
log_regr = LogisticRegression()

# fit the model 
log_regr.fit(X_train, y_train)

# evaluate the model 
print(log_regr.score(X_test, y_test)) # accuracy


# prediction 
y_predicted = log_regr.predict(X_test)
print()
print(y_predicted)

print()
print(y_test)
```

```python
print(log_regr.coef_)
print()
print(log_regr.intercept_)
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
#sns.pairplot(iris, hue='species', kind='scatter')
sns.pairplot(iris, hue='species', kind='kde')
```

```python

sns.barplot(x = y_predicted, y = y_test)
plt.xlabel('X-test')
plt.ylabel('y-test')
plt.title('Iris Seaborn line plot')
```

### Conclusion:

# Random Forest

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)

# create the Model
model = RandomForestClassifier(n_estimators = 10, random_state=1)

# train the Model
model.fit(X_train, y_train)

# make prediciton
y_predicted = model.predict(X_test)
print(y_predicted)

# evaluate Model
model.score(X_test, y_test) # accuracy
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_predicted)
sns.heatmap(cm, annot=True)

plt.xlabel('Truth')
plt.ylabel('y_predicted')
```

```python
metric = metrics.classification_report(y_test, y_predicted)
print('Classification Matrix:')
print('metric: ', metric)
```

### Conclusion: our random forest model showed 96% accuracy, precision and recall are also relatively high, which means we can use it for future.
